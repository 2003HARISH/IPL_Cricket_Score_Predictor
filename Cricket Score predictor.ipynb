# Test/train split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,[7,8,9,12,13]].values
y = df.iloc[:, 14].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=102)
```

```python
X_train
```

# scaling


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

# creating a model


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
```

```python

from tensorflow.keras.optimizers import Adam

# Create a sequential model
model = Sequential()

# Add a dense layer with input dimension 5 (number of features)
model.add(Dense(units=19, activation='relu', input_dim=5))
  # Adding dropout layer with 20% dropout rate
model.add(Dense(units=19, activation='relu'))

model.add(Dense(units=19, activation='relu'))

model.add(Dense(units=19, activation='relu'))



# Output layer (linear activation for regression)
model.add(Dense(units=1))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['mae','mse'])
```

```python
model.summary()
```

```python
X_train.shape
```

# Training the data

```python
model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
          batch_size=128,epochs=400)
```

# Evaluating the model

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.columns
```

```python
losses[['loss','val_loss']].plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
explained_variance_score(y_test,)
```

```python
predictions = model.predict(X_test)
```
